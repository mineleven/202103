```
import requests
```

```
url = 'https://opendata.cwb.gov.tw/api/v1/rest/datastore/A-B0062-001?Authorization=CWB-FADFD658-6B05-4C7D-9682-54D6375D0AB1&locationName=%E8%87%BA%E6%9D%B1%E7%B8%A3'
```

```
def retrieval_api(api_url):
  response = requests.get(api_url)
  # print('Status Code: ' + str(response.status_code))
  return str(response.json())
```

```
location_json = retrieval_api(url)
```

```
print(location_json)
```

```
print(type(location_json))
```

```
location_json = location_json.replace("\'", "\"")
print(location_json)
```

```
import json
location_dic = json.loads(location_json)
print(type(location_dic))
print(location_dic)
```

```
location = location_dic['records']['locations']['location'][0]['locationName']
date = location_dic['records']['locations']['location'][0]['time'][75]['dataTime']
sunrise = location_dic['records']['locations']['location'][0]['time'][75]['parameter'][1]['parameterName']
sunriseTime = location_dic['records']['locations']['location'][0]['time'][75]['parameter'][1]['parameterValue']
sunset = location_dic['records']['locations']['location'][0]['time'][75]['parameter'][5]['parameterName']
sunsetTime = location_dic['records']['locations']['location'][0]['time'][75]['parameter'][5]['parameterValue']
```
